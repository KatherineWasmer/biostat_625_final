```python
data = pd.read_csv("cleaned_data_v1.csv")
```

```python
# filter data where systolic bp is less than diastolic (not possible) 
sys_lt_dias = data[data["ap_hi"] < data["ap_lo"]]
```

```python
sys_lt_dias["ap_hi"]
```

```python
# replace with NaN  
sys_lt_dias["ap_hi"] = np.nan
```

```python
sys_lt_dias["ap_lo"] = np.nan
```

```python
sys_lt_dias["BP Category"] = np.nan
```

```python
sys_lt_dias.head()
```

```python
sys_lt_dias = sys_lt_dias.iloc[:, 1:]
```

```python
sys_lt_dias
```

```python
data = data.iloc[:, 1:] 
```

```python
def filter_and_replace_outliers(full_df, subset):
    ids_to_remove = np.array(subset["id"]) 
    data_filtered = full_df[~full_df.id.isin(ids_to_remove)]
    # concatenate filtered data w/subset and reset idx 
    data = pd.concat([data_filtered, subset])
    data = data.reset_index()
    data.drop("index", axis=1, inplace=True)
    return data 
```

```python
data = filter_and_replace_outliers(data, sys_lt_dias)
```

```python
data
```

```python
# subset data where systolic > 300
sys_outliers = data[data["ap_hi"] > 300] 
```

```python
# set impossible values to null 
sys_outliers["ap_hi"] = np.nan 
```

```python
data = filter_and_replace_outliers(data, sys_outliers)
```

```python
data
```

```python
# subset data where diastolic < 30 
dias_outliers = data[data["ap_lo"] < 30] 
```

```python
dias_outliers["ap_lo"] = np.nan 
data = filter_and_replace_outliers(data, dias_outliers)
data
```

```python
print(data.info())
```

```python
def remove_bp_categories_for_nulls(sys, dias, bp_category):
    if np.isnan(sys) or np.isnan(dias):
        return np.nan
    return bp_category
```

```python
data["BP Category"] = data.apply(lambda x: remove_bp_categories_for_nulls(x.ap_hi, x.ap_lo, x["BP Category"]), axis=1)
```
